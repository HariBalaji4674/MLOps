## List All Files From Directory or System

```python
### List All Files in The Given Directory

import os

path = "."

for filename in os.listdir(path):
    print(filename)
    
```

```python
## Get User name 

import getpass

username = getpass.getuser()
print(username)

### Get User name using PWD ond OS Modules

import os
import pwd

username1 = pwd.getpwuid(os.getuid()).pw_name
print(username1)
```

```python
## Find and Print All Running Process in a system for hari user

import subprocess
import pwd
import os

## Find the user

user = pwd.getpwuid(os.getuid()).pw_name

ps_run = subprocess.run(['ps', '-u', user],capture_output=True,text=True)
print(ps_run.stdout)
```
