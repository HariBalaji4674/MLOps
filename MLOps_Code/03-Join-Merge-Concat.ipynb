```python
import pandas as pd

df1 = pd.DataFrame({
    'A': ['A0','A1','A2','A3','A4','A5'],
    'B': ['B0','B1','B2','B3','B4','B5'],
    'C': ['C0','C1','C2','C3','C4','C5'],
},index = [0,1,2,3,4,5])

df1
```

```python
df2 = pd.DataFrame({
    'A': ['A6','A7','A8','A9','A10','A11'],
    'B': ['B6','B7','B8','B9','B10','B11'],
    'C': ['C6','C7','C8','C9','C10','C11'],
},index = [6,7,8,9,10,11])

df2
```

```python
df3 = pd.DataFrame({
    'A': ['A12','A13','A14','A15','A16','A17'],
    'B': ['B12','B13','B14','B15','B16','B17'],
    'C': ['C12','C13','C14','C15','C16','C17'],
},index = [12,13,14,15,16,17])

df3
```

```python
from re import A


pd.concat([df1,df2,df3],axis=1)
```

```python
from numpy import inner


pd.merge(df1,df2,how='inner')
```

```python
left = pd.DataFrame({
'A':['A1','A2','A3'],
'B':['B1','B2','B3'],
'key':['k0','k1','k2']
})

left
```

```python
right = pd.DataFrame({
'A1':['A4','A5','A6'],
'B1':['B4','B5','B6'],
'key1':['k0','k1','k2']
})

right
```

```python
left.join(right,how='outer')
```
